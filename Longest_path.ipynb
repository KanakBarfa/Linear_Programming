```python
from pyomo.environ import *
import math

n = 20

column_index = range(n)
row_index = range(n)
model = ConcreteModel()
```

```python
dist_list =[[1,2,463],
[1,3,256],
[2,5,373],
[2,4,172],
[3,20,956],
[3,4,565],
[4,18,774],
[18,20,652],
[18,6,596],
[20,19,265],
[19,8,679],
[5,6,362],
[6,7,448],
[6,17,410],
[7,8,420],
[7,13,553],
[8,9,314],
[9,10,444],
[10,12,223],
[10,11,649],
[10,13,200],
[11,12,271],
[11,14,831],
[12,13,261],
[13,14,436],
[14,15,663],
[15,16,474],
[16,17,666]]

for i in range(len(dist_list)):
  dist_list[i][0] = dist_list[i][0] - 1
  dist_list[i][1] = dist_list[i][1] - 1

new = []

for l in dist_list:
  temp = [l[1], l[0], l[2]]

  new.append(temp)

dist_list = dist_list + new
print(dist_list)
print(len(dist_list))
```

```python
path = [[0 for p in range(n)] for q in range(n)]
distance = [[0 for p in range(n)] for q in range(n)]
for ind in dist_list:
	path[ind[0]][ind[1]] = 1
	distance[ind[0]][ind[1]] = ind[2]

print(path)
print(distance)
```

```python
model.xij = Var(row_index, column_index, domain = Binary)
```

```python
model.u = Var(range(n),domain= NonNegativeIntegers)
```

```python
model.obj = Objective(expr = sum(distance[i][j]*model.xij[i,j] for i in row_index for j in column_index), sense = maximize)
```

```python
model.constraints = ConstraintList()

for i in row_index:
  for j in column_index:
    model.constraints.add(expr = model.xij[i,j] <= path[i][j])
```

```python
model.constraints.add(model.u[0] == 0)
```

```python
for j in column_index:
  model.constraints.add(expr = sum(model.xij[i,j] for i in row_index if i != j) == sum(model.xij[j,i] for i in column_index if i !=j))

for i in row_index:
  model.constraints.add(expr = sum(model.xij[i,j] for j in column_index if i != j)<=1)
```

```python
for i in row_index:
  for j in column_index:
    model.constraints.add(expr = model.xij[i,j]+model.xij[j,i] <= 1)
```

```python


for i in range(1,n):
  for j in range(1,n):
    if i!=j:
      model.constraints.add(expr = (model.u[i]-model.u[j]+n*model.xij[i,j]) <= n-1)

```

```python
model.pprint()
```

```python
opt = SolverFactory('cbc')
```

```python
result = opt.solve(model, tee = True)
print('Solver status:', result.solver.status)
print('Solver termination condition:',result.solver.termination_condition)
```
